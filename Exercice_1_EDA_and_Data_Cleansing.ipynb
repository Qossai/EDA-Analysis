```python
import seaborn as sns
import matplotlib.pyplot as plt

categorical_nominal_columns = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']

for column in categorical_nominal_columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=column, data=df)
    plt.title(f'Count Plot of {column}')
    plt.show()
```

```python
#Similar to the above
categorical_nominal_columns = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']
for column in categorical_nominal_columns:
  sns.countplot(x=column, data = df, color = "green")
  plt.show()
```

```python
categorical_ordinal = ['Education', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'RelationshipSatisfaction', 'StockOptionLevel']
#For categorical-ordinal variables, I willuse box plots

for column in categorical_ordinal:
  sns.boxplot(data = df, color = "black", x = column,y = "Attrition", order= df[column].unique().sort())
  plt.show()
```

```python
continuous = ['Unnamed: 0', 'Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating',
              'StandardHours', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager',
              'SalaryAddition', 'YearlyHollidayDays', 'DrivingKilometers']

for column in continuous:
  sns.histplot(df[column], kde = True)
  plt.show()
```

```python
# converting categorical to numirical

df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
df['BusinessTravel'] = df['BusinessTravel'].map({'Travel_Rarely': 0, 'Travel_Frequently': 1, 'Travel_Very_Frequently': 2})
df['Department'] = df['Department'].map({'Sales': 0, 'Research & Development': 1, 'Human Resources': 2, 'Marketing': 3})
df['EducationField'] = df['EducationField'].map({'Life Sciences': 0, 'Other': 1, 'Marketing': 2, 'Technical Degree': 3})
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df['JobRole'] = df['JobRole'].map({'Healthcare Representative': 0, 'Sales Executive': 1, 'Technical Support Engineer': 2, 'Business Analyst': 3, 'Research Scientist': 4, 'Sales Representative': 5, 'Manager': 6})
df['MaritalStatus'] = df['MaritalStatus'].map({'Married': 0, 'Single': 1, 'Divorced': 2, 'Widowed': 3})
df['Over18'] = df['Over18'].map({'Yes': 1, 'No': 0})
df['OverTime'] = df['OverTime'].map({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
#boxplots to detect which variables have extreme values (outliers).
continuous = ['Unnamed: 0', 'Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating',
              'StandardHours', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager',
              'SalaryAddition', 'YearlyHollidayDays', 'DrivingKilometers']
for column in continuous:
  plt.figure(figsize = (10,4))
  sns.boxplot(x = df[column])
  plt.title('Boxplot of ' + column)
  plt.show()
```

```python
# Generate a correlation matrix using the 'corr()' method.
corr_matrix = df.corr()
# Visualize the correlation matrix using Seaborn's heatmap function.
plt.figure(figsize=(25, 25))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```
